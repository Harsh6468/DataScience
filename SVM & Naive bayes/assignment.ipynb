## **SVM & Naive Bayes Assignment**

### **Theory**

### **1. What is a Support Vector Machine (SVM)?**
SVM is a supervised machine learning algorithm used for classification and regression tasks. It finds the hyperplane that best separates the data into different classes by maximizing the margin between the closest data points (support vectors) of different classes.

### **2. What is the difference between Hard Margin and Soft Margin SVM?**
| Feature               | Hard Margin SVM            | Soft Margin SVM                     |
| --------------------- | -------------------------- | ----------------------------------- |
| Misclassification     | Not allowed                | Allowed with penalty (C)            |
| Slack Variables       | Not used                   | Used ($\xi_i$)                      |
| Flexibility           | Rigid (sensitive to noise) | Flexible (robust to noise/outliers) |
| Dataset Compatibility | Linearly separable only    | Works on overlapping data           |


### **3. What is the mathematical intuition behind SVM?**
* **Goal**: Find a hyperplane that **maximizes the margin** between classes.
* **Margin**: Distance between the hyperplane and the **closest points** (support vectors).
* **Hyperplane equation**:

  $$
  w \cdot x + b = 0
  $$
* **Optimization Objective**:

  $$
  \min \frac{1}{2} \|w\|^2 \quad \text{subject to } y_i(w \cdot x_i + b) \geq 1
  $$
* **Soft Margin**: Adds slack variables $\xi_i$ and a penalty term $C \sum \xi_i$ to allow some misclassification.
* **Dual Form**: Uses Lagrange multipliers and **kernel trick** to handle non-linear data.
* **Prediction**:

  $$
  f(x) = \text{sign}\left( \sum_i \alpha_i y_i K(x_i, x) + b \right)
  $$


### **4. What is the role of Lagrange Multipliers in SVM?**
Lagrange Multipliers help solve the constrained optimization problem in SVM. They are used to form the dual problem which simplifies computation and enables the use of kernel functions.

### **5. What are Support Vectors in SVM?**
Support vectors are the data points that lie closest to the decision boundary or hyperplane. These are the most critical elements in the dataset, as they directly affect the position and orientation of the hyperplane. The SVM model is entirely defined by these support vectors, and removing them would change the decision boundary.

### **6.  What is a Support Vector Classifier (SVC)?**
A Support Vector Classifier is the classification version of the SVM algorithm. It constructs a hyperplane in a high-dimensional space to separate different classes. SVC can use different kernel functions to handle non-linear separations in the data.

### **7. What is a Support Vector Regressor (SVR)?**
SVR is the regression counterpart of SVM. Instead of trying to find a hyperplane that separates classes, SVR tries to fit the best line (or function) within a threshold (epsilon) to predict continuous values. The goal is to keep predictions within a certain error margin while being as flat as possible.

### **8. What is the Kernel Trick in SVM?**
The kernel trick allows SVM to solve non-linear problems by implicitly mapping the input features into high-dimensional space without computing the transformation explicitly. Common kernels include Linear, Polynomial, and Radial Basis Function (RBF).

### **9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.**
| Feature                  | **Linear Kernel**                    | **Polynomial Kernel**                                     | **RBF Kernel (Radial Basis Function)**                |
| ------------------------ | ------------------------------------ | --------------------------------------------------------- | ----------------------------------------------------- |
| **Definition**           | $K(x, x') = x \cdot x'$              | $K(x, x') = (x \cdot x' + c)^d$                           | $K(x, x') = \exp(-\gamma \|x - x'\|^2)$               |
| **Type**                 | Linear                               | Non-linear                                                | Non-linear                                            |
| **Use Case**             | Linearly separable data              | Data with polynomial relationships                        | Complex, non-linear relationships                     |
| **Hyperparameters**      | None                                 | Degree $d$, constant $c$                                  | Gamma $\gamma$                                        |
| **Computation**          | Fast                                 | Moderate                                                  | Expensive on large datasets                           |
| **Overfitting Risk**     | Low                                  | Increases with degree                                     | High if gamma is too large                            |
| **Flexibility**          | Low                                  | Medium (depends on degree)                                | High                                                  |
| **Interpretability**     | Easy                                 | Harder                                                    | Hard                                                  |
| **Typical Applications** | Text classification, linear problems | Image recognition, problems with polynomial-like patterns | Image classification, time-series, speech recognition |


### **10. What is the effect of the C parameter in SVM?**
The C parameter controls the trade-off between achieving a low error on the training data and minimizing the margin. A small C makes the margin larger but allows more misclassifications (higher bias), while a large C focuses on classifying all training examples correctly (lower bias, higher variance).

### **11. What is the role of the Gamma parameter in RBF Kernel SVM?**
Gamma defines how far the influence of a single training example reaches. A low gamma means far-reaching influence, resulting in smoother decision boundaries. A high gamma makes the model focus more on individual data points, which can lead to overfitting.

### **12. What is the Naïve Bayes classifier, and why is it called "Naïve"?**
Naïve Bayes is a **probabilistic classification algorithm** based on **Bayes' Theorem**. It is commonly used for **text classification**, spam detection, sentiment analysis, and more.

It calculates the probability that a data point belongs to a particular class based on the **prior probability** of the class and the **likelihood** of the features, assuming they are **independent**.

#### Bayes' Theorem:

$$
P(C | X) = \frac{P(X | C) \cdot P(C)}{P(X)}
$$

Where:

* $P(C | X)$: Posterior probability of class $C$ given input $X$
* $P(X | C)$: Likelihood of input $X$ given class $C$
* $P(C)$: Prior probability of class $C$
* $P(X)$: Evidence (normalizing constant)

It is called 'naïve' because it assumes that all features are conditionally independent given the class label, which is rarely true in real datasets. Despite this, it often performs well in practice.

This means:
$$
P(x_1, x_2, ..., x_n | C) = \prod_{i=1}^{n} P(x_i | C)
$$


### **13. What is Bayes’ Theorem?**

$$
P(C | X) = \frac{P(X | C) \cdot P(C)}{P(X)}
$$

Where:

* $P(C | X)$: Posterior probability of class $C$ given input $X$
* $P(X | C)$: Likelihood of input $X$ given class $C$
* $P(C)$: Prior probability of class $C$
* $P(X)$: Evidence (normalizing constant)

### **14. Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes.**
Naïve Bayes has **three main variants**, each suited for **different types of data**:

1. **Gaussian Naïve Bayes**

* **Used For**: **Continuous numeric features**.
* **Assumes**: Features follow a **normal (Gaussian) distribution** within each class.
* **Formula**:

  $$
  P(x_i|y) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x_i - \mu)^2}{2\sigma^2}\right)
  $$
* **Example Use Case**: Classifying **medical records**, **sensor data**, or any data with continuous measurements.

2. **Multinomial Naïve Bayes**

* **Used For**: **Discrete counts**, such as **word frequencies** in documents.
* **Assumes**: Features represent **counts** (non-negative integers).
* **Common In**: **Text classification** problems (spam detection, news categorization).
* **Formula**:

  $$
  P(x|y) = \frac{(\sum_i x_i)!}{\prod_i x_i!} \prod_i \theta_{yi}^{x_i}
  $$

  where $\theta_{yi}$ is the probability of word $i$ in class $y$.

3. **Bernoulli Naïve Bayes**

* **Used For**: **Binary features** (0 or 1), indicating presence or absence.
* **Assumes**: Each feature is **Boolean** — either the word occurs or not.
* **Common In**: Text classification where presence is more important than frequency.
* **Formula**:

  $$
  P(x|y) = \prod_i P(x_i = 1|y)^{x_i} \cdot P(x_i = 0|y)^{1 - x_i}
  $$

### **15. When should you use Gaussian Naïve Bayes over other variants?**
Gaussian Naïve Bayes is preferred when the features are continuous and follow a normal distribution. It's suitable for problems like classification of sensor data or any data modeled by Gaussian distributions.

### **16. What are the key assumptions made by Naïve Bayes?**
1. All features are independent given the class label.
2. The presence or absence of a feature does not depend on any other feature.
3. The distribution of features (Gaussian, Multinomial, Bernoulli) matches the variant of Naïve Bayes used.

### **17. What are the advantages and disadvantages of Naïve Bayes?**
**Advantages**:
- Fast and efficient
- Performs well with high-dimensional data
- Works well with text classification

**Disadvantages**:
- Strong independence assumption rarely holds
- May not perform well with correlated features

### **18. Why is Naïve Bayes a good choice for text classification?**
Naïve Bayes works well in high-dimensional spaces like text data, where word frequencies are used as features. The independence assumption is reasonable in this context, and training is extremely fast even on large corpora.

### **19. Compare SVM and Naïve Bayes for classification tasks.**
- **SVM**:
  - Works well for both linear and non-linear problems using kernels
  - Sensitive to parameter tuning
  - Computationally expensive for large datasets

- **Naïve Bayes**:
  - Simple, fast, scalable
  - Works well for text and high-dimensional data
  - Assumes feature independence, which may not hold

### **20. How does Laplace Smoothing help in Naïve Bayes?**
Laplace Smoothing helps handle zero-probability problems in Naïve Bayes by adding a small constant (usually 1) to each count. This ensures that no probability is ever exactly zero, which would otherwise completely rule out unseen words or features in predictions.

```python
# 24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary:
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

x, y = make_classification(n_samples=100, n_features=2, n_redundant=0, n_clusters_per_class=1, random_state=42)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

model = SVC(kernel='poly')
model.fit(x_train, y_train)

h = 0.02
x_min, x_max = x_test[:, 0].min() - 1, x_test[:, 0].max() + 1
y_min, y_max = x_test[:, 1].min() - 1, x_test[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
z = model.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)
plt.contourf(xx, yy, z, alpha=0.8)
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test, edgecolors='k')
plt.title("SVM with Polynomial Kernel")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# 27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

x, y = make_classification(n_samples=200, n_features=2, n_redundant=0, n_clusters_per_class=1, flip_y=0.1, class_sep=1.0, random_state=42)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

C_values = [0.01, 0.1, 1, 10, 100]

plt.figure(figsize=(15, 10))
for i, C in enumerate(C_values):
    clf = SVC(kernel='linear', C=C)
    clf.fit(x_train, y_train)
    
    plt.subplot(2, 3, i + 1)
    ax = plt.gca()
    
    x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
    y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300), np.linspace(y_min, y_max, 300))
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.title(f"C = {C}")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")

plt.suptitle("SVM Decision Boundaries for Different C Values", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# 30. Write a Python program to train a Gaussian Naïve Bayes model and compare the predictions before and after Laplace Smoothing.
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

gnb = GaussianNB()
gnb.fit(X_train, y_train)
pred_default = gnb.predict(X_test)
acc_default = accuracy_score(y_test, pred_default)

X_train_smooth = X_train + np.random.normal(0, 0.01, X_train.shape)
X_test_smooth = X_test + np.random.normal(0, 0.01, X_test.shape)

gnb_smooth = GaussianNB()
gnb_smooth.fit(X_train_smooth, y_train)
pred_smooth = gnb_smooth.predict(X_test_smooth)
acc_smooth = accuracy_score(y_test, pred_smooth)

print("Original Accuracy:", acc_default)
print("Smoothed Accuracy (simulated):", acc_smooth)
```

```python
# 45. Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC score.
from sklearn.datasets import load_breast_cancer
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

# Load binary classification dataset
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

# Train Naive Bayes
model = GaussianNB()
model.fit(X_train, y_train)
y_probs = model.predict_proba(X_test)[:, 1]

# Compute ROC-AUC
roc_auc = roc_auc_score(y_test, y_probs)
print("ROC-AUC Score:", roc_auc)
```
